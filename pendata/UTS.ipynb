```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Pemahaman Data (Data Understanding)

## 1.1. Load Data

```python
df = pd.read_csv('cirrhosis.csv')

print("5 Data Pertama:")
display(df.head())
```

## 1.2. Eksplorasi Dasar

```python
# Info dasar
print("\nInformasi Dataset:")
print(df.info())
```

```python
# Statistik deskriptif
print("\nStatistik Deskriptif:")
display(df.describe(include='all'))
```

```python
# Cek missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

## 1.3. Visualisasi Distribusi Target

```python
sns.countplot(x='Status', data=df)
plt.title('Distribusi Target')
plt.show()
```

## 1.4. Visualisasi Distribusi Fitur Numerik

```python
df.hist(bins=30, figsize=(15, 10))
plt.suptitle("Distribusi Fitur Numerik")
plt.show()
```

# 2. Preprocessing Data

## 2.1. Encoding Variabel Kategorikal
Berdasarkan output df.isnull().sum(), jika ada missing values:

```python
categorical_cols = ['Sex', 'Ascites', 'Hepatomegaly', 'Spiders', 'Edema', 'Drug', 'Stage']
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
df['Status'] = df['Status'].map({'D': 0, 'C': 1})
```

## 2.2. Handle Missing Values

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

## 2.3. Korelasi antar fitur

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Korelasi antar fitur")
plt.show()
```

## 2.4. Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

# Pisahkan fitur & target
X = df.drop(['ID', 'Status'], axis=1)
y = df['Status']

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## 2.5. Split Data 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

## 2.6. Training Model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

## 2.7. Evaluasi

```python
print("Accuracy:", model.score(X_test, y_test))

from sklearn.metrics import roc_auc_score
y_proba = model.predict_proba(X_test)[:, 1]
print("AUC-ROC:", roc_auc_score(y_test, y_proba))
```

# 3. Pemodelan (Klasifikasi)

## 3.1. Pilih Model
Kita akan coba 3 model untuk perbandingan:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

# Inisialisasi model
models = {
  'Logistic Regression': LogisticRegression(max_iter=1000),
  'KNN': KNeighborsClassifier(n_neighbors=5),
  'Random Forest': RandomForestClassifier()
}
```

## 3.2. Training & Prediksi

```python
from sklearn.metrics import accuracy_score, classification_report

for name, model in models.items():
  # Training
  model.fit(X_train, y_train)

  # Prediksi
  y_pred = model.predict(X_test)

  # Evaluasi
  print(f"\n{name} Performance:")
  print("Accuracy:", accuracy_score(y_test, y_pred))
  print(classification_report(y_test, y_pred))
```

## 3.3. Validasi Silang (Cross-Validation)

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(RandomForestClassifier(), X_scaled, y, cv=5)
print("\nCross-Validation Scores:", scores)
print("Average CV Score:", scores.mean())
```

# 4. Evaluasi Model

## 4.1. Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['D', 'C'])  #  Ubah label encoder manual

plt.figure(figsize=(6, 4))
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

## 4.2. Feature Importance (Random Forest)

```python
importances = model.feature_importances_
feature_names = X.columns

plt.figure(figsize=(10,6))
sns.barplot(x=importances, y=feature_names)
plt.title('Feature Importance')
plt.show()
```
