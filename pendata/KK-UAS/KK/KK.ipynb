## Data Preparation

````python
print("Kolom yang tersedia di df_clean:\n", df_clean.columns.tolist())
````

### Data Cleaning

````python
df_clean = df_clean.drop(columns=["Tanggal Didaftar", "NIK"])

# Handling Missing Values
print("Missing Values Sebelum Cleaning:")
print(df_clean.isnull().sum())
````

````python
# Daftar kolom yang akan diimputasi
kategorikal_cols = ["Status HIV", "Riwayat Diabetes Melitus", "Lokasi Anatomi", "Hasil Pemeriksaan Foto Toraks", "Jenis Kelamin"]
numerik_cols = ["Umur", "Tahun"]

# Imputasi kolom kategorikal dengan modus
for col in kategorikal_cols:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])

# Imputasi kolom numerik dengan median
for col in numerik_cols:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna(df_clean[col].median())

print("\nMissing Values Setelah Cleaning:")
print(df_clean.isnull().sum())
````

````python
# Simpan versi sebelum hapus duplikat
df_before = df_clean.copy()
print("\nJumlah Duplikat Sebelum:", df_before.duplicated().sum())

# Hapus duplikat dan simpan hasilnya ke variabel baru
df_clean = df_before.drop_duplicates()
print("Jumlah Duplikat Sesudah:", df_clean.duplicated().sum())
````

### Data Reduction

````python
print("Usia maksimal di data:", df_clean["Umur"].max())
print("Usia minimal di data:", df_clean["Umur"].min())
````

````python
LOWER_BOUND = 1    # Bayi baru lahir (0 tahun) dianggap tidak relevan untuk data TBC
UPPER_BOUND = 90   # Usia >90 tahun dianggap outlier ekstrem

outliers_lower = df_clean[df_clean["Umur"] < LOWER_BOUND]
outliers_upper = df_clean[df_clean["Umur"] > UPPER_BOUND]
total_outliers = len(outliers_lower) + len(outliers_upper)

# Hapus outlier
df_clean = df_clean[(df_clean["Umur"] >= LOWER_BOUND) & (df_clean["Umur"] <= UPPER_BOUND)]

print("\n=== LAPORAN OUTLIER USIA ===")
print(f"Batas valid: {LOWER_BOUND}-{UPPER_BOUND} tahun")
print(f"Jumlah outlier usia < {LOWER_BOUND}: {len(outliers_lower)}")
print(f"Jumlah outlier usia > {UPPER_BOUND}: {len(outliers_upper)}")
print(f"Total outlier dihapus: {total_outliers}")
print("\nContoh data outlier:")
print(pd.concat([outliers_lower.head(2), outliers_upper.head(2)])[["Umur", "Tahun"]])
````

````python
plt.figure(figsize=(15, 6))

# Before
plt.subplot(1, 2, 1)
sns.histplot(df_before["Umur"], bins=range(0,101,5), kde=False, color='red')
plt.axvspan(0, LOWER_BOUND, color='red', alpha=0.3, label='Outlier Bawah')
plt.axvspan(UPPER_BOUND, 100, color='orange', alpha=0.3, label='Outlier Atas')
plt.title('Distribusi Usia Sebelum Outlier Treatment', fontweight='bold')
plt.xlabel('Usia (tahun)')
plt.xticks(range(0,101,10))
plt.legend()

# After
plt.subplot(1, 2, 2)
sns.histplot(df_clean["Umur"], bins=range(LOWER_BOUND, UPPER_BOUND+1,5), kde=False, color='green')
plt.title('Distribusi Usia Sesudah Outlier Treatment', fontweight='bold')
plt.xlabel('Usia (tahun)')
plt.xticks(range(LOWER_BOUND, UPPER_BOUND+1,10))

plt.tight_layout()
plt.show()
````

### Data Transformation

````python
### 1. Encoding Data Kategorikal ###
from sklearn.preprocessing import LabelEncoder

# Kolom untuk Label Encoding (kategorikal biner)
label_cols = ["Jenis Kelamin", "Riwayat Diabetes Melitus"]
le = LabelEncoder()
for col in label_cols:
    df_clean[col] = le.fit_transform(df_clean[col])

# Kolom untuk One-Hot Encoding (kategorikal multi-kelas)
ohe_cols = ["Lokasi Anatomi", "Hasil Pemeriksaan Foto Toraks"]
df_clean = pd.get_dummies(df_clean, columns=ohe_cols, drop_first=True)

# Pisahkan fitur (X) dan target (y)
X = df_clean.drop(columns=["Status HIV"])
y = df_clean["Status HIV"]

# Cek distribusi kelas target sebelum ADASYN
print("\nDistribusi Kelas Target Sebelum ADASYN:")
print(y.value_counts())
````

````python
### 2. Handling Class Imbalance ###
from imblearn.over_sampling import ADASYN

# Lakukan oversampling dengan ADASYN
adasyn = ADASYN(random_state=42)
X_resampled, y_resampled = adasyn.fit_resample(X, y)

# Cek distribusi kelas setelah ADASYN
print("\nDistribusi Kelas Target Sesudah ADASYN:")
print(pd.Series(y_resampled).value_counts())
````

````python
### 3. Normalisasi/Standarisasi ###
from sklearn.preprocessing import StandardScaler

# Normalisasi dengan Z-Score (StandardScaler)
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X_resampled)
````

### Data Splitting

````python
from sklearn.model_selection import train_test_split

# Bagi data menjadi training (80%) dan testing (20%)
X_train, X_test, y_train, y_test = train_test_split(
    X_normalized, 
    y_resampled, 
    test_size=0.2, 
    random_state=42,
    stratify=y_resampled  # Untuk menjaga distribusi kelas
)

print("\n=== DATA SPLITTING ===")
print(f"Jumlah Data Training: {X_train.shape[0]}")
print(f"Jumlah Data Testing: {X_test.shape[0]}")
print(f"Shape X_train: {X_train.shape}")
print(f"Shape X_test: {X_test.shape}")
````

## Modeling

````python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Inisialisasi model
models = {
    "SVM": SVC(kernel='linear'),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(n_estimators=100),
}
````

````python
# Dictionary untuk menyimpan hasil evaluasi
results = {}

# Loop untuk training dan evaluasi model
for name, model in models.items():
    # Training model
    model.fit(X_train, y_train)
    
    # Prediksi
    y_pred = model.predict(X_test)
    
    # Evaluasi
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True)
    
    # Simpan hasil
    results[name] = {
        "accuracy": accuracy,
        "precision": report['weighted avg']['precision'],
        "recall": report['weighted avg']['recall'],
        "f1": report['weighted avg']['f1-score']
    }
    
    # Print hasil
    print(f"\n=== {name} ===")
    print(f"Akurasi: {accuracy:.4f}")
    print(classification_report(y_test, y_pred))
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Negatif', 'Positif'], 
                yticklabels=['Negatif', 'Positif'])
    plt.title(f'Confusion Matrix - {name}')
    plt.ylabel('Aktual')
    plt.xlabel('Prediksi')
    plt.show()
````

````python
# --------------------------
# Hyperparameter Tuning
# --------------------------
from sklearn.model_selection import GridSearchCV

# Parameter grid
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}

# simplify
# param_grid = {
#     'n_estimators': [100],
#     'max_depth': [10, 20]
# }

# Grid Search
grid_search = GridSearchCV(
    estimator=RandomForestClassifier(),
    param_grid=param_grid,
    # n_iter=10,  # jumlah kombinasi yang akan dicoba
    cv=5,
    scoring='f1_weighted'
)

grid_search.fit(X_train, y_train)

best_rf = grid_search.best_estimator_
y_pred_rf = best_rf.predict(X_test)

print("\n=== HASIL TUNING RANDOM FOREST ===")
print(f"Parameter Terbaik: {grid_search.best_params_}")
print(classification_report(y_test, y_pred_rf))

import joblib

joblib.dump(best_rf, 'best_random_forest_model.pkl')
print("\nModel terbaik telah disimpan sebagai 'best_random_forest_model.pkl'")

results_df = pd.DataFrame(results).T
print("\n=== PERBANDINGAN MODEL ===")
print(results_df.sort_values(by='f1', ascending=False))
````
